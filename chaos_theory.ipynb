```python
from numpy import*
from pylab import * # for plotting commands
from scipy.integrate import odeint
```

```python
def deriv(y,t): # return derivatives of the array y
   a = -1.0
   b = -1.0
   return array([ y[1], a*y[0]+b*y[1] ])

time = linspace(0.0,10,2000) # Time array: (t_start, t_end, Nt)

yinit = array([0.2,0.0])   # Initial values for y[0] and y[1]
y1 = odeint(deriv,yinit,time)  # Call the ode solver odeint( );

yinit = array([0.5,0.0])   # Initial values for y[0] and y[1]
y2 = odeint(deriv,yinit,time)  # Call the ode solver odeint( );

yinit = array([1.0,0.0])   # Initial values for y[0] and y[1]
y3 = odeint(deriv,yinit,time)  # Call the ode solver odeint( );


figure()
plot(time, y1[:,0], '-r', label='i.c : (0.2, 0)')
plot(time, y2[:,0], '-c', label='i.c : (0.5, 0)')
plot(time, y3[:,0], '-m', label='i.c : (1.0, 0)')
legend()
grid()
xlabel("t")
ylabel("y")
savefig("Coupled_ODE_Solution.pdf")
show()

figure()
plot(y1[:,0],y1[:,1], '-r', label='x-v phase space')
plot(y2[:,0],y2[:,1], '-c' )
plot(y3[:,0],y3[:,1], '-m' )
legend()
grid()
xlabel("x")
ylabel("v")
savefig("PhaseSpace_Coupled_ODE_Solution.pdf")
show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import * # for plotting commands
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
#####################################################################
# mapping of Lorentz to Python variables
# y[0]=y1; y[1] = y2, y[2]=y3
#
def deriv(y,t,s,r,b): # return derivatives of the array y
   dydt=[  s*y[1]-s*y[0], -y[0]*y[2] + r*y[0] - y[1], \
           y[0]*y[1] - b*y[2] ]
   return dydt

r=28.0
s=10.0
b=8./3.
time = linspace(0.0,50,2000) # Time array: (t_start, t_end, Nt)
yinit = array([0.5, 0.0, 0.0 ])   # Initial values for y[0], y[1], y[2]
y = odeint(deriv, yinit, time, args=(s,r,b))  # Call the ode solver odeint( );

figure()
plot(y[:,0],y[:,1], '-', label='x-y Lorentz')
legend()
grid()
xlabel("x ")
ylabel("y ")
savefig("xy_Lorenz.pdf")
show()

figure()
plot(y[:,1],y[:,2], '-', label='y-z Lorentz')
legend()
grid()
xlabel("y ")
ylabel("z ")
savefig("yz_Lorenz.pdf")
show()

figure()
plot(time,y[:,0], '-', label='x vs time')
legend()
grid()
xlabel("t ")
ylabel("x ")
savefig("xt_Lorenz.pdf")
show()

figure()
plot(time,y[:,1], '-', label='y vs time')
legend()
grid()
xlabel("t ")
ylabel("y ")
savefig("yt_Lorenz.pdf")
show()

figure()
plot(time,y[:,2], '-', label='z vs time')
legend()
grid()
xlabel("t ")
ylabel("z ")
savefig("zt_Lorenz.pdf")
show()


points = np.array([y[:,0],y[:,1],y[:,2]])

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot(y[:,0],y[:,1],y[:,2], marker = '.')
ax.scatter(*points.T[0], color = 'red')
plt.show()
```

```python
from numpy import*
from pylab import * # for plotting commands
from scipy.integrate import odeint
#####################################################################
def deriv(y,t): # return derivatives of the array y
   eps=1.1
   return array([ y[1], -y[0]+eps*(1.0-y[0]*y[0])*y[1] ])

time = linspace(0.0,10,4000) # Time array: (t_start, t_end, Nt)

yinit = array([2.1,0.0])   # Initial values for y[0] and y[1]
y1 = odeint(deriv,yinit,time)  # Call the ode solver odeint( );

yinit = array([2.1,-2.0])   # Initial values for y[0] and y[1]
y2 = odeint(deriv,yinit,time)  # Call the ode solver odeint( );

yinit = array([0.1,0.0])   # Initial values for y[0] and y[1]
y3 = odeint(deriv,yinit,time)  # Call the ode solver odeint( );

yinit = array([3.1,2.0])   # Initial values for y[0] and y[1]
y4 = odeint(deriv,yinit,time)  # Call the ode solver odeint( );

yinit = array([-1,0.0])   # Initial values for y[0] and y[1]
y5 = odeint(deriv,yinit,time)  # Call the ode solver odeint( );



figure()
plot(time, y1[:,0], '-', label='Van der Pol Oscillator - Numeric Solution')

legend()
grid()
xlabel("t")
ylabel("y")
savefig("VanderPol_Oscillator_by_Python.pdf")
show()

figure()
plot(y1[:,0],y1[:,1], '-k', label='2.1, 0')
plot(y2[:,0],y2[:,1], '-b', label='2.1, 2')
plot(y3[:,0],y3[:,1], '-c', label='0.0, 0')
plot(y4[:,0],y4[:,1], '-m', label='3.1, 2')
plot(y5[:,0],y5[:,1], '-r', label='-1, 0')

legend()
grid()
xlabel("x")
ylabel("v")
savefig("VanderPol_PhaseSpace.pdf")
show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
k = 8.99e9  # Coulomb's constant (N m^2 / C^2)
q = 1e-9  # point charge (C)

# Grid of points
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Electric field components
Ex = k * q * X / (X**2 + Y**2)**(3/2)
Ey = k * q * Y / (X**2 + Y**2)**(3/2)

# Plotting the electric field
plt.figure(figsize=(10, 5))
plt.quiver(X, Y, Ex, Ey, color='b')
plt.title('Electric Field of a Point Charge')
plt.xlabel('X (m)')
plt.ylabel('Y (m)')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Initial conditions
v0 = 50  # initial velocity (m/s)
theta = 45  # launch angle (degrees)
g = 9.81  # acceleration due to gravity (m/s^2)

# Convert angle to radians
theta_rad = np.radians(theta)

# Time of flight
t_flight = 2 * v0 * np.sin(theta_rad) / g

# Time points
t = np.linspace(0, t_flight, num=500)

# Trajectory equations
x = v0 * np.cos(theta_rad) * t
y = v0 * np.sin(theta_rad) * t - 0.5 * g * t**2

# Plotting the trajectory
plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.title('Projectile Motion')
plt.xlabel('Distance (m)')
plt.ylabel('Height (m)')
plt.ylim(bottom=0)
plt.grid()
plt.show()
```
